# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus

````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression safely."""
    try:
        # Evaluate mathematical string
        return str(eval(expression))
    except ZeroDivisionError:
        return "Error: Division by zero"
    except Exception:
        return "Error in calculation"
````

````python
import re
from typing import List, Dict, Any

# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract unique keywords (length > 4) from text, stripping out punctuation and preserving order."""
    try:
        # Remove punctuation
        clean_text = re.sub(r'[^\w\s]', '', text)
        words = clean_text.split()

        # Maintain order while ensuring uniqueness
        keywords_dict = {}
        for w in words:
            w_lower = w.lower()
            if len(w_lower) > 4:
                keywords_dict[w_lower] = None

        return list(keywords_dict.keys())[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (IMPLEMENTED)

def agent(query: str) -> Dict[str, Any]:
    query_lower = query.lower()

    try:
        # Route 1: Math / Calculation queries
        if "calculate" in query_lower:
            expression = re.sub(r"calculate", "", query_lower, flags=re.IGNORECASE).strip()
            expression = re.sub(r"[^0-9+\-*/().\s]", "", expression)

            if not expression:
                return {"type": "error", "result": "No valid expression found to calculate."}

            calc_result = calculator(expression)
            if calc_result.startswith("Error"):
                return {"type": "error", "result": calc_result}

            return {"type": "calculation", "result": calc_result}

        # Route 2: Keyword extraction queries
        elif "keywords" in query_lower:
            text = re.sub(r"extract keywords from", "", query, flags=re.IGNORECASE).strip()
            text = re.sub(r"keywords in", "", text, flags=re.IGNORECASE).strip()

            if not text:
                return {"type": "error", "result": "No text provided for keyword extraction."}

            keywords = extract_keywords(text)
            return {"type": "keywords", "result": keywords}

        # Route 3: General queries — fallback response
        else:
            return {
                "type": "general",
                "result": f"This is a general query: '{query}'. No specific tool was needed."
            }

    except Exception as e:
        return {"type": "error", "result": f"Something went wrong: {str(e)}"}
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````
